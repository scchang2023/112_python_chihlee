```python
#標準寫法
import matplotlib.pyplot as plt

# 1.製作figure
fig = plt.figure()

# 2.在1.的figure上製作一個axes
ax_1 = fig.add_subplot(1, 1, 1)

# 3.將圖的數據儲存在axes中
X = [0, 1, 2]
Y = [3, 4, 5]
ax_1.plot(X, Y)

# 4.顯示圖
plt.show()
```

```python
fig = plt.figure()
ax_1 = fig.add_subplot(1,3,1)
ax_1.set_title("axes 1")

ax_2 = fig.add_subplot(1,3,2)
ax_2.set_title("axes 2")

ax_3 = fig.add_subplot(1,3,3)
ax_3.set_title("axes 3")
```

```python
student1 = {'chinese':78, 'english':80, 'math':92, 'history':75, 'discover':85}
student2 = {'chinese':90, 'english':75, 'math':98, 'history':65, 'discover':75}
student3 = {'chinese':50, 'english':105, 'math':60, 'history':45, 'discover':80}

subjects =list(student1.keys())
value1 =list(student1.values())
value2 = list(student2.values())
value3 = list(student3.values())

figure = plt.figure(figsize=(8.0, 5.0), dpi=80.0, facecolor="#cccccc")
ax = figure.add_subplot()
ax.plot(subjects, value1, 'r')
ax.plot(subjects, value2, 'g')
ax.plot(subjects, value3, 'b')

plt.show()
```
